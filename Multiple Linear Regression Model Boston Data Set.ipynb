### Multiple Linear Regression Model:

#### Importing data:

```r
library(MASS)
data = Boston
```

#### Dividing train set and test set:

```r
id = sample(1:nrow(data),round(nrow(data)*0.6))
train_set = data[id,]
test_set = data[-id,]
```

### Training the model:

```r
model = lm(medv~.,data=train_set)
```

#### Running the model on the test set and ploting predicted results vs real results:

```r
predicted = predict(model,test_set)
plot(predicted,test_set$medv,xlab="Predicted",ylab="Real data",main="Multiple Linear Regression Model")
```
