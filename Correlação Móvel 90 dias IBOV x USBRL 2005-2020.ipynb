# Vou plotar aqui a correlação móvel entre IBOV e USDBRL:

```python
#importamos os módulos necessários:
from pandas_datareader import data as wb 
import pandas as pd 
import matplotlib.pyplot as plt 
```

# Importando as bases: 

```python
#Importamos dados de USDBRL e Ibovespa:
data=pd.DataFrame()
data['USDBRL'] = wb.DataReader('USDBRL=X',data_source='yahoo',start='2005-01-01')['Adj Close']
data['IBOV'] = wb.DataReader('^BVSP',data_source='yahoo',start='2005-01-01')['Adj Close']

#Retiramos todas as linha que contêm nans da base:
data.dropna(inplace=True)
```

# Fazendo e plotando a correlação móvel em janelas de 90 dias:

```python
#Calculo os retornos diários da base: 
data_daily_returns = data.pct_change()

#Calculo a correlação móvel dos preços com janelas de 90 dias desde 2005 e em seguida plotamos:
correlacao_movel_precos = data['USDBRL'].rolling(90).corr(data['IBOV'])
plt.figure(figsize=(16,4))
plt.plot(correlacao_movel_precos)
plt.ylabel('Correlação')
plt.xlabel('Data')
plt.title('Preços: Correlação em janelas de 90 dias')
plt.show()

#Calculo a correlação móvel dos retornos diários com janelas de 90 dias desde 2005 e em seguida plotamos:
correlacao_movel_precos = data_daily_returns['USDBRL'].rolling(90).corr(data_daily_returns['IBOV'])
plt.figure(figsize=(16,4))
plt.plot(correlacao_movel_precos)
plt.ylabel('Correlação')
plt.xlabel('Data')
plt.title('Retornos: Correlação em janelas de 90 dias')
plt.show()
```
