# Vou mostrar como pegar dados do BACEN no Quandl: 

# Aqui vou importar dados de juros de vários países desde 2000: 

```python
import quandl
import pandas as pd 

paises = ['Australia','Canada','China','Zona Euro','EUA','Japão','México','Inglaterra','Rússia']
tickers_bacen = {'Australia':"BCB/17880",'Canada':"BCB/17881",'China':"BCB/17899",'Zona Euro':"BCB/17900",'EUA':"BCB/18152"\
                ,'Japão':"BCB/17903",'México':"BCB/17904",'Inglaterra':"BCB/17908",'Rússia':"BCB/17905"}

df_juros = pd.DataFrame()

for i in paises: 
    df_juros[i]=quandl.get(tickers_bacen[i],start_date='2000-01-01')['Value']
```
